```python
def count_Factors(n):
    listOfFactors = []
    for i in range(1,n+1):
        if n%i == 0:
            listOfFactors.append(i)
    return len(listOfFactors)

def get_dictionary():
    listOfTriangleNumber = []
    dictionaryWithFactors = {}
    for i in range(1,500):
            sumOfJ = 0
            for j in range(1,i+1):
                sumOfJ = sumOfJ + j
            listOfTriangleNumber.append(sumOfJ)
            dictionaryWithFactors[sumOfJ] = count_Factors(sumOfJ)
    return dictionaryWithFactors

dictionaryWithFactors = get_dictionary()

t = int(input())
for repeat in range(t):
    n = int(input())
    N = n
    #print(dictionaryWithFactors)
    for i in dictionaryWithFactors:
        #print("i = " + str(i) + " " +str(dictionaryWithFactors[i]))
        if dictionaryWithFactors[i] > N:
            print(i)
            break
    
```
