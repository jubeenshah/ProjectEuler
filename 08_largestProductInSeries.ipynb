```python
#!/bin/python3

import sys


t = int(input("Enter Number of test cases less than 10000 : " ).strip())
if t >= 1 and t <= 10000:
    for a0 in range(t):
        n,k = input("Enter Number of Digits in the current test case with number of adjacent digits : ").strip().split(' ')
        n,k = [int(n),int(k)]
        num = input("Enter the " + str(n) + " digit Number! : ").strip()
        start = 0
        end = k
        listOfProduct = []
        for i in range (0, (n-k)+1):
            prod = 1
            for j in num[start:end]:
                prod = prod * int(j)
            listOfProduct.append(prod)
            start = start + 1
            end = end + 1
        print()
        print("Answer : " + str(max(listOfProduct)))
else:
    print("Please respect constraints")
```
